```python
poses_path = '/home/eghaleb/Projects/ViTPose/vis_results/*.pkl'
all_poses_dir = '/home/eghaleb/Projects/gestural_alignment/data/final_poses/'


poses_path = glob.glob(poses_path)
for i, path in enumerate(poses_path):
    print(i, path)
       # save the landmarks
    subject_name = path.split('/')[-1]
    subject_name = subject_name.replace('.pkl', '.npy')
    speaker = subject_name.split('.')[0][-1]
    subject_ID = subject_name.split('.')[0].split('_')[1]
    save_path = all_poses_dir + subject_name

     # check if the file already exists
    if os.path.exists(save_path):
       
        print('File already exists:', save_path)
        continue
   
    with open(path, 'rb') as f:
        poses = pickle.load(f)
    main_subject_landmarks = np.zeros((len(poses), 133, 3))

    def get_main_subject_ID(poses):
        # get the pose with the highest score
        poses_scores = []
        for pose in poses:
            score = pose['keypoints'][:,2]
            poses_scores.append(np.sum(score))
        return np.argmax(poses_scores)
    
    for i, pose in enumerate(poses):
        if len(pose) > 1:
            main_subject_ID = get_main_subject_ID(pose)
            main_subject_landmarks[i] = pose[main_subject_ID]['keypoints']
        else:
            if len(pose) == 1:
                main_subject_landmarks[i] = pose[0]['keypoints']
            else:
                main_subject_landmarks[i] = np.zeros((133, 3))
                print('No pose detected in frame', i)
 
    np.save(save_path, main_subject_landmarks)
```

```python
poses_path = '../data/final_poses/*.npy'
modified_poses_path = '../data/selected_poses/'
import glob
import numpy as np
selected_points = np.array([0, 5, 6, 7, 8, 9, 10, 91, 95, 96, 99, 100, 103, 104, 107, 108, 111, 112, 116, 117, 120, 121, 124, 125, 128, 129, 132])
# read all poses

for path in glob.glob(poses_path):
    print(path)
    # check if the file already exists
    if os.path.exists(modified_poses_path + path.split('/')[-1]):
        print('File already exists:', modified_poses_path + path.split('/')[-1])
        continue
    poses = np.load(path)
    # select poses
    selected_poses = poses[:, selected_points, :]
    # save selected poses
    np.save(modified_poses_path + path.split('/')[-1], selected_poses)
    print('Saved', modified_poses_path + path.split('/')[-1])
```
